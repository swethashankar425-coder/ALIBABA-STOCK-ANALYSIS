## Residual Analysis

Residual analysis is a critical step in time series modeling to validate the assumptions of the model and check if the residuals behave like white noise. If the residuals are white noise (i.e., uncorrelated, zero mean, constant variance), it suggests that the model has captured all the available information in the series.

We will perform the following checks for both ARIMA and SARIMAX model residuals:

1.  **Residual Plot:** To visually check for any patterns over time.
2.  **Histogram/KDE Plot:** To check if residuals are normally distributed around zero.
3.  **ACF/PACF Plots:** To detect any remaining autocorrelation in the residuals.
4.  **Ljung-Box Test:** A formal statistical test for white noise (no autocorrelation).

First, we need to ensure the `train_data` and `test` sets are correctly defined for the 'Open' price, as these were used for fitting the previous models, and then re-fit the models.

```
# Re-define train_data and test from 'Open' price to be consistent with previous model fitting
# Assuming `data` DataFrame is already loaded and indexed by Date

# Split data into training and testing sets
train_size = int(len(data) * 0.8)
train_data = data['Open'].iloc[:train_size]
test = data['Open'].iloc[train_size:]

print(f"Train data length: {len(train_data)}")
print(f"Test data length: {len(test)}")
```

### ARIMA Model Residuals Analysis

```
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.stats.diagnostic import acorr_ljungbox
import matplotlib.pyplot as plt
import seaborn as sns # Import seaborn for histplot

# Re-fit the ARIMA model to get the model_fit_arima object
# Using previously determined best-fit orders for ARIMA: (7, 1, 7)
# This is a re-run of the code from cell GaZ_qbu-ZLJN for residual extraction

model_arima = ARIMA(train_data, order=(7, 1, 7))
model_fit_arima = model_arima.fit()

# Extract residuals
residuals_arima = model_fit_arima.resid

print("ARIMA Model Summary:")
print(model_fit_arima.summary())

# Plot residuals over time
plt.figure(figsize=(14, 6))
plt.plot(residuals_arima)
plt.title('ARIMA Model Residuals Over Time')
plt.xlabel('Date')
plt.ylabel('Residual')
plt.grid(True)
plt.show()

# Plot histogram and KDE of residuals
plt.figure(figsize=(10, 6))
sns.histplot(residuals_arima, kde=True)
plt.title('Histogram and KDE of ARIMA Model Residuals')
plt.xlabel('Residual')
plt.ylabel('Frequency')
plt.show()

# Plot ACF and PACF of residuals
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))
plot_acf(residuals_arima, lags=40, ax=ax1, title='ACF of ARIMA Model Residuals')
plot_pacf(residuals_arima, lags=40, ax=ax2, title='PACF of ARIMA Model Residuals')
plt.tight_layout()
plt.show()

# Ljung-Box test for autocorrelation
ljung_box_results_arima = acorr_ljungbox(residuals_arima, lags=[10, 20, 30])
print("Ljung-Box Test Results for ARIMA Model Residuals:")
print(ljung_box_results_arima)
```

## Project Report Summary: Stock Data Analysis and Forecasting

### 1. Introduction and Project Goal

This project aimed to analyze historical stock data, specifically for Alibaba (`Ali_Baba_Stock_Data.csv`), to understand its patterns and build time series models for forecasting. The primary goal was to explore the data, identify trends and seasonality, and then implement and evaluate ARIMA and SARIMAX models for stock price prediction.

### 2. Data Loading and Initial Exploration

*   **Data Source:** The project utilized the `Ali_Baba_Stock_Data.csv` dataset, which includes daily stock information such as 'Open', 'High', 'Low', 'Close', 'Adj Close', and 'Volume'.
*   **Data Quality:** Initial checks confirmed no missing values or duplicate rows, indicating a clean dataset for analysis.
*   **Data Type Conversion:** The 'Date' column was successfully converted to datetime objects and set as the DataFrame's index, which is a crucial step for time series analysis.

### 3. Exploratory Data Analysis (EDA)

*   **Overall Trends:** Visualizations of 'Open', 'High', and 'Low' prices over time revealed general trends and fluctuations in the stock's value.
*   **Date Filtering:** Specific time windows were examined (e.g., 2020-2021) to observe stock behavior during particular periods.
*   **Time Resampling:** The data was resampled to annual (Year-End) and monthly frequencies to observe long-term minimums, maximums, and mean trends for 'Open' and 'High' prices.
*   **Moving Averages:** 30-day, 50-day, and 100-day Simple Moving Averages (SMA) were calculated and plotted for 'Open' and 'Close' prices. These smoothed out short-term noise, highlighting underlying trends and providing insights into momentum.
*   **ACF/PACF Plots:** Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) plots were generated for 'Adj Close' price. These plots indicated significant autocorrelation, suggesting that time series models are appropriate, and provided initial guidance for selecting AR and MA orders.

### 4. Stationarity Testing

*   **Augmented Dickey-Fuller (ADF) Test:** An ADF test was performed on the 'Adj Close' price series. The high p-value (0.594) led to the conclusion that the series is **non-stationary**. This finding confirmed the necessity of differencing (setting 'd' or 'D' > 0) when building ARIMA/SARIMAX models to achieve stationarity.

### 5. ARIMA Model Implementation and Residual Analysis

*   **Model Building:** An ARIMA model with `order=(7, 1, 7)` was fitted to the `train_data` (80% of 'Open' prices), and predictions were generated for the `test` set.
*   **Prediction Plot:** The actual vs. predicted prices showed the model's ability to capture the general trend.
*   **Residual Analysis:**
    *   **Normality:** The Jarque-Bera test (p-value = 0.00) indicated that residuals are **not normally distributed**, which is common for financial time series.
    *   **Autocorrelation:** The Ljung-Box test results showed **no significant autocorrelation** in the residuals at lags 1, 10, 20, or 30 (high p-values). This is a strong indicator that the ARIMA model effectively captured the linear dependencies in the data.
    *   **Heteroskedasticity:** Significant **heteroskedasticity** (p-value = 0.00) was observed in the residuals, meaning the variance of errors changes over time. This suggests that while the mean structure is well-modeled, the volatility is not fully captured.

### 6. SARIMAX Model Implementation and Residual Analysis

*   **Model Building:** A SARIMAX model with `non_seasonal_order=(4, 1, 4)` and `seasonal_order=(7, 1, 7, 5)` (seasonal period `s=5` for weekly trading days) was fitted. Initial parameter selection required adjustments to avoid overlap errors.
*   **Prediction Plot:** Similar to ARIMA, the SARIMAX model generated predictions that followed the general trend of the stock prices.
*   **Residual Analysis:**
    *   **Normality & Heteroskedasticity:** Consistent with ARIMA, SARIMAX residuals were also **not normally distributed** and exhibited significant **heteroskedasticity**.
    *   **Autocorrelation:** The Ljung-Box test for SARIMAX residuals revealed **significant autocorrelation remaining** at higher lags (p-values of 0.004 at lag 10, 0.011 at lag 20, and 0.031 at lag 30). This is a critical finding, indicating that the current SARIMAX model has **not fully captured all the seasonal or long-term dependencies** in the data.

### 7. Conclusion and Next Steps

*   **ARIMA Performance:** The ARIMA model proved effective in removing linear autocorrelation from the series. However, its inability to model non-normal, heteroskedastic residuals points towards limitations in capturing volatility.
*   **SARIMAX Limitations:** Despite its seasonal component, the SARIMAX model showed room for improvement due to remaining autocorrelation in its residuals at certain lags. This suggests that the chosen orders (`p,d,q,P,D,Q,s`) might not be optimal.
